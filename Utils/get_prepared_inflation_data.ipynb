# Подготовка данных по инфляции
Приведение ежемесячных показателей к квартальным и создание временного ряда по инфляции

(соотношение инфляции к предыдущему временному периоду)

###### Источник данных: https://www.statbureau.org/ru/russia/inflation-tables

```python
import pandas as pd
```

```python
df = pd.read_csv("../Data/data_input/russia.inflation.monthly (statbureau.org).csv")
df.set_index("Год", inplace=True)

# исходные данные
# отображают размер инфляции за месяц
df
```

```python
# приведение показателей инфляции к коэффициенту инфляции 
# для возможности вычисления квартальных значений

df.fillna(1, inplace=True)

if " Всего" in df.columns:
    total = df[" Всего"].copy()
    del df[" Всего"]
    df /= 100
    df += 1

df["q1"] = df[[" Январь", " Февраль", " Март"]].prod(axis="columns")
df["q2"] = df[[" Апрель", " Май", " Июнь"]].prod(axis="columns")
df["q3"] = df[[" Июль", " Август", " Сентябрь"]].prod(axis="columns")
df["q4"] = df[[" Октябрь", " Ноябрь", " Декабрь"]].prod(axis="columns")
```

```python
inf_koeff = df[["q1", "q2", "q3", "q4"]]
inf_koeff
```

```python
# создание единого временного ряда 
# квартальные значения инфляции за временные периоды 1993 - 2023

inf_r = pd.melt(inf_koeff, ignore_index=False)

quarter_dict = {
    "q1": "01.01",
    "q2": "04.01",
    "q3": "07.01",
    "q4": "10.01"
}

inf_r["Quarter_num"] = inf_r.variable.map(quarter_dict)

inf_r.reset_index(inplace=True)
inf_r["Date"] = inf_r["Год"].astype("str") + "." + inf_r.Quarter_num
inf_r["Date"] = pd.to_datetime(inf_r["Date"])
inf_r.sort_values("Date", inplace=True)

inf_r = inf_r[["value", "Date"]].rename(columns={"value": "Value"})
inf_r = inf_r[(inf_r["Date"] >= "1993-01-01") & (inf_r["Date"] < "2023-07-01")]
inf_r.set_index("Date", inplace=True)
inf_r.tail()
```

```python
# поверхностная визулализация временного ряда инфляции
inf_r.plot()
```
